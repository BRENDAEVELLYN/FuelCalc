# Geração de Dataset de Preços de Combustíveis

## Descrição do Código

Este script em Python gera um dataset que simula os preços da gasolina e etanol no Brasil entre 2010 e 2028. O código realiza as seguintes etapas:

1. **Definição de Parâmetros:**
   - Define um intervalo de anos de 2010 a 2028.
   - Gera valores aleatórios para inflação (entre 1% e 5%) e preços do petróleo (entre R$50 e R$100).

2. **Cálculo dos Preços:**
   - Calcula o preço da gasolina e do etanol com base em uma relação linear com a inflação e o preço do petróleo, adicionando um ruído aleatório para simular a variação real dos preços.

3. **Criação do DataFrame:**
   - Cria um DataFrame contendo as colunas: 
     - **Ano**
     - **Inflação (%)**
     - **Preço Gasolina (R$/L)**
     - **Preço Etanol (R$/L)**
     - **Preço do Petróleo (R$/L)**
     - **Taxa de Câmbio (R$)** (gerada aleatoriamente entre R$4 e R$5).

4. **Exportação para CSV:**
   - O DataFrame é salvo em um arquivo CSV chamado `dataset_precos_combustiveis.csv`.

Esse dataset pode ser utilizado para análises estatísticas e modelagem de preços de combustíveis.


```python
import pandas as pd
import numpy as np

# Definindo parâmetros
anos = np.arange(2010, 2029)  # De 2010 a 2028
inflacao = np.random.uniform(1.0, 5.0, len(anos))  # Inflação aleatória entre 1% e 5%
preco_petroleo = np.random.uniform(50, 100, len(anos))  # Preço do petróleo entre R$50 e R$100

# Gerando dados de preço da gasolina e etanol com base em uma relação com inflação e preço do petróleo
preco_gasolina = 3 + 0.1 * inflacao + 0.05 * preco_petroleo + np.random.normal(0, 0.2, len(anos))
preco_etanol = 2 + 0.09 * inflacao + 0.04 * preco_petroleo + np.random.normal(0, 0.2, len(anos))

# Criando DataFrame
data = pd.DataFrame({
    'Ano': anos,
    'Inflação (%)': inflacao,
    'Preço Gasolina (R$/L)': preco_gasolina,
    'Preço Etanol (R$/L)': preco_etanol,
    'Preço do Petróleo (R$/L)': preco_petroleo,
    'Taxa de Câmbio (R$)': np.random.uniform(4, 5, len(anos))  # Taxa de câmbio entre R$4 e R$5
})

# Salvar o dataset em um arquivo CSV
data.to_csv('dataset_precos_combustiveis.csv', index=False)
print(data.head())
```

# Interpretação do Mapa de Calor da Correlação

## 1. Valores de Correlação
- **Escala de Valores**: Os valores de correlação variam de -1 a 1:
  - **1**: Correlação positiva perfeita (as variáveis se movem na mesma direção).
  - **-1**: Correlação negativa perfeita (as variáveis se movem em direções opostas).
  - **0**: Sem correlação (nenhuma relação linear entre as variáveis).
- **Valores Próximos a 1 ou -1**: Indicam uma forte relação entre as variáveis. Valores acima de 0,5 ou abaixo de -0,5 geralmente são considerados significativos.

## 2. Cores
- **Cores Quentes (vermelho)**: Representam correlações positivas altas.
- **Cores Frias (azul)**: Representam correlações negativas altas.
- **Cores Neutras (branco ou cinza)**: Indicam correlações fracas ou inexistentes.

## 3. Interpretação dos Resultados
- **Examine as Relações**: Olhe para as correlações mais altas ou mais baixas entre as variáveis:
  - Por exemplo, se o **Preço Gasolina (R$/L)** tem uma correlação de **0,8** com o **Preço do Petróleo (R$/L)**, isso sugere que um aumento no preço do petróleo está associado a um aumento no preço da gasolina.
  - Se a **Inflação (%)** tem uma correlação de **0,7** com o **Preço Etanol (R$/L)**, isso indica que a inflação tende a aumentar juntamente com o preço do etanol.
- **Identifique Relações Inesperadas**: Às vezes, você pode encontrar correlações inesperadas que podem levar a novas perguntas ou investigações sobre a relação entre as variáveis.

## 4. Limitações
- **Correlação não Implica Causalidade**: Uma correlação alta não significa que uma variável causa a outra. Pode haver outros fatores não considerados que influenciam essas variáveis.
- **Variáveis Omissas**: Pode haver variáveis não incluídas no seu dataset que afetam as correlações observadas.

## 5. Exemplo de Interpretação
Se você observar que:
- **Preço Gasolina (R$/L)** e **Preço do Petróleo (R$/L)** têm uma correlação de **0,85**:
  - Isso sugere que, quando o preço do petróleo sobe, o preço da gasolina tende a subir também.
  
- **Inflação (%)** e **Taxa de Câmbio (R$)** têm uma correlação de **-0,4**:
  - Isso pode indicar que, à medida que a inflação aumenta, a taxa de câmbio tende a diminuir (ou vice-versa), mas essa relação não é muito forte.

# Treinamento de Modelo de Regressão Linear para Previsão de Preços de Combustíveis

## Descrição do Código

Este script em Python utiliza o scikit-learn para treinar um modelo de regressão linear, com o objetivo de prever os preços da gasolina com base em variáveis econômicas. O código realiza as seguintes etapas:

### 1. Importação de Bibliotecas
- Importa as bibliotecas necessárias, incluindo `train_test_split`, `LinearRegression`, `mean_squared_error` e `r2_score` do módulo `sklearn`.

### 2. Carregamento do Dataset
- Carrega um conjunto de dados de preços de combustíveis a partir de um arquivo CSV chamado `dataset_precos_combustiveis.csv`.

### 3. Definição das Variáveis
- Define as variáveis independentes (X) como:
  - **Ano**
  - **Inflação (%)**
  - **Preço do Petróleo (R$/L)**
  - **Taxa de Câmbio (R$)**
- Define a variável dependente (y) como:
  - **Preço Gasolina (R$/L)**

### 4. Divisão dos Dados
- Divide o dataset em conjuntos de treinamento e teste, utilizando 30% dos dados para teste e 70% para treinamento, garantindo a reprodutibilidade com um `random_state` definido.

### 5. Criação e Treinamento do Modelo
- Cria uma instância do modelo de regressão linear e o treina utilizando o conjunto de dados de treinamento.

### 6. Fazendo Previsões
- Utiliza o modelo treinado para fazer previsões sobre os preços da gasolina com base no conjunto de teste.

### 7. Avaliação do Modelo
- Calcula o Erro Médio Quadrático (MSE) e o coeficiente de determinação (R²) para avaliar a performance do modelo, imprimindo os resultados.

Esse código permite avaliar a eficácia do modelo de regressão linear na previsão dos preços da gasolina, ajudando a entender a relação entre as variáveis econômicas e o preço dos combustíveis.



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Carregar o Dataset
data = pd.read_csv('dataset_precos_combustiveis.csv')

# Definir as variáveis independentes e dependentes
X = data[['Ano', 'Inflação (%)', 'Preço do Petróleo (R$/L)', 'Taxa de Câmbio (R$)']]
y = data['Preço Gasolina (R$/L)']

# Dividir os dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e treinar o modelo de regressão linear
model = LinearRegression()
model.fit(X_train, y_train)

# Fazer previsões
y_pred = model.predict(X_test)

# Avaliar o modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Erro Médio Quadrático: {mse}')
print(f'R²: {r2}')
```

# Acurácia

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Carregar o Dataset
data = pd.read_csv('dataset_precos_combustiveis.csv')

# Definir as variáveis independentes e dependentes
X = data[['Ano', 'Inflação (%)', 'Preço do Petróleo (R$/L)', 'Taxa de Câmbio (R$)']]
y = data['Preço Gasolina (R$/L)']

# Dividir os dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e treinar o modelo de regressão linear
model = LinearRegression()
model.fit(X_train, y_train)

# Fazer previsões
y_pred = model.predict(X_test)

# Avaliar o modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Calcular a acurácia relativa
acuracia_relativa = 1 - (np.sum(np.abs(y_test - y_pred)) / np.sum(np.abs(y_test)))

# Exibir os resultados
print(f'Erro Médio Quadrático: {mse}')
print(f'R²: {r2}')
print(f'Acurácia Relativa: {acuracia_relativa * 100:.2f}%')
```

# Streamlit APP

```python
import streamlit as st
import pandas as pd
from sklearn.linear_model import LinearRegression

# Carregar o dataset de preços
data = pd.read_csv('dataset_precos_combustiveis.csv')

# Criar o modelo de regressão linear para a previsão
X = data[['Ano', 'Inflação (%)', 'Preço do Petróleo (R$/L)', 'Taxa de Câmbio (R$)']]
y_gasolina = data['Preço Gasolina (R$/L)']
y_etanol = data['Preço Etanol (R$/L)']

# Treinar o modelo de regressão linear para gasolina
model_gasolina = LinearRegression()
model_gasolina.fit(X, y_gasolina)

# Treinar o modelo de regressão linear para etanol
model_etanol = LinearRegression()
model_etanol.fit(X, y_etanol)

# Função para calcular o preço futuro baseado na inflação
def prever_preco_futuro(inflacao, ano):
    entrada = [[ano, inflacao, data['Preço do Petróleo (R$/L)'].mean(), data['Taxa de Câmbio (R$)'].mean()]]
    preco_gasolina = model_gasolina.predict(entrada)
    preco_etanol = model_etanol.predict(entrada)
    return preco_gasolina[0], preco_etanol[0]

# Configurações do Streamlit
st.title("Calculadora de Economia de Combustível")

# Página principal
menu = st.sidebar.selectbox("Escolha uma página", ["Calculadora de Combustível", "Previsão de Preço Futuro"])

if menu == "Calculadora de Combustível":
    st.header("Calculadora: Etanol ou Gasolina?")
    
    # Entrada do usuário
    preco_gasolina = st.number_input("Preço da Gasolina (R$/L):", min_value=0.0, step=0.01)
    preco_etanol = st.number_input("Preço do Etanol (R$/L):", min_value=0.0, step=0.01)
    consumo_gasolina = st.number_input("Consumo de Gasolina (Km/L):", min_value=0.0, step=0.01)
    consumo_etanol = st.number_input("Consumo de Etanol (Km/L):", min_value=0.0, step=0.01)
    
    if st.button("Calcular"):
        # Calcular o custo por km
        custo_km_gasolina = preco_gasolina / consumo_gasolina
        custo_km_etanol = preco_etanol / consumo_etanol
        
        st.write(f"Custo por Km com Gasolina: R$ {custo_km_gasolina:.2f}")
        st.write(f"Custo por Km com Etanol: R$ {custo_km_etanol:.2f}")
        
        if custo_km_gasolina < custo_km_etanol:
            st.success("Gasolina compensa mais!")
        else:
            st.success("Etanol compensa mais!")
            
elif menu == "Previsão de Preço Futuro":
    st.header("Previsão de Preço Futuro com Base na Inflação")
    
    # Entrada para previsão
    ano = st.number_input("Insira o ano futuro para previsão:", min_value=2024, step=1)
    inflacao_futura = st.slider("Insira a inflação estimada (%):", min_value=0.0, max_value=20.0, step=0.1)
    
    if st.button("Prever Preços"):
        preco_gasolina_futuro, preco_etanol_futuro = prever_preco_futuro(inflacao_futura, ano)
        
        st.write(f"Previsão de preço da Gasolina em {ano}: R$ {preco_gasolina_futuro:.2f}/L")
        st.write(f"Previsão de preço do Etanol em {ano}: R$ {preco_etanol_futuro:.2f}/L")
```
